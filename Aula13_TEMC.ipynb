```python
from pyswarm import pso
import numpy as np
import matplotlib.pyplot as plt

interval = [-100, 100]

x = np.linspace(interval[0], interval[1], 500)
y = np.array([griewank_function([xi, 0]) for xi in x])

plt.plot(x, y)

lb = [interval[0], interval[0]]
ub = [interval[1], interval[1]]

xopt,fopt = pso(griewank_function, lb,ub)
```
