Exercício 1 GradientBoosting Classifier + DecisionTree


```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
dados = load_breast_cancer()
X = dados.data
y = dados.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
clf_ens = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)

clf_ens.fit(X_train, y_train)

y_pred_ens = clf_ens.predict(X_test)

acc_ens = accuracy_score(y_test,y_pred_ens)
```

```python
clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

acc = accuracy_score(y_test,y_pred)
```

```python
path = '/content/sample_data/BostonHousing.csv'
```

```python
import pandas as pd

dados = pd.read_csv(path)
```

```python
X = dados.drop('medv', axis=1)
y = dados['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
r1 = MLPRegressor(random_state=1, max_iter=500)
r2 = LinearRegression()
r3 = SVR()
```

```python
r_ens = VotingRegressor(estimators=[('mlpRegressor', r1), ('linearRegression', r2), ('svr', r3)])

r_ens.fit(X_train,y_train)

y_pred_ans = r_ens.predict(X_test)

mse = mean_squared_error(y_test, y_pred_ans)
```
