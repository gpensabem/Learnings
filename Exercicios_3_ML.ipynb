```python
import sklearn
```

```python
from sklearn.datasets import load_wine
from sklearn.feature_selection import SelectKBest
#no caso de regress√£o precisa importar o f_regression
from sklearn.feature_selection import f_classif
```

```python
wine = load_wine()
```

```python
X = wine.data
y = wine.target

target_names = wine.target_names
feature_names = wine.feature_names
```

```python
X
```

```python
selector = SelectKBest(f_classif, k=3)
X_new=selector.fit_transform(X,y)
X_new.shape
```

```python
from sklearn import decomposition
from sklearn import datasets
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)
X_pca
```

```python
import matplotlib.pyplot as plt
scatter=plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, alpha=.8, lw=2)

plt.xlabel('Componente 1 expressa '+ str("{:.2f}".format(100*pca.explained_variance_ratio_[0])) + '% de variabilidade')
plt.ylabel('Componente 2 expressa ' + str("{:.2f}".format(100*pca.explained_variance_ratio_[1])) + '% de variabilidade')
plt.legend(scatter.legend_elements()[0],target_names, loc="lower right", title="Classes")
plt.title('PCA da base de dados WINE')
```
