```python
from sklearn.datasets import load_iris, load_breast_cancer
```

```python
iris = load_iris()
cancer = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_iris = iris.data
y_iris = iris.target
```

```python
X_cancer = cancer.data
y_cancer = cancer.target
```

```python
X_iris_train, X_iris_test, y_iris_train, y_iris_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)
```

```python
X_cancer_train, X_cancer_test, y_cancer_train, y_cancer_test = train_test_split(X_cancer, y_cancer, test_size=0.2, random_state=42)
```

```python
X_iris_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn import metrics

def Gaussian_Naive_Bayes(X_train, y_train, X_test, y_test):
  gnb = GaussianNB()
  gnb.fit(X_train, y_train)
  y_pred = gnb.predict(X_test)
  acc = metrics.accuracy_score(y_test, y_pred)
  f1 = metrics.f1_score(y_test, y_pred, average='macro')

  print(f"Acc: {acc} F1-Score: {f1}")
```

```python
Gaussian_Naive_Bayes(X_iris_train, y_iris_train, X_iris_test, y_iris_test)
Gaussian_Naive_Bayes(X_cancer_train, y_cancer_train, X_cancer_test, y_cancer_test)
```

```python
def Bernoulli_Naive_Bayes(X_train, y_train, X_test, y_test):
  bnb = BernoulliNB()
  bnb.fit(X_train, y_train)
  y_pred = bnb.predict(X_test)
  acc = metrics.accuracy_score(y_test, y_pred)
  f1 = metrics.f1_score(y_test, y_pred, average='macro')

  print(f"Acc: {acc} F1-Score: {f1}")
```

```python
Bernoulli_Naive_Bayes(X_iris_train, y_iris_train, X_iris_test, y_iris_test)
Bernoulli_Naive_Bayes(X_cancer_train, y_cancer_train, X_cancer_test, y_cancer_test)
```
