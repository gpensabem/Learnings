```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from google.colab import drive
drive.mount('/content/drive')
# Carregar uma imagem em escala de cinza
image = plt.imread('/content/drive/MyDrive/Aprendizados/img1.jpg')
if image.ndim == 3:
    image = image[:, :, 0]  # Caso seja uma imagem de cor, a transforma em escalaz de cinza

# Executar a decomposição SVD
U, S, VT = svd(image)

# Selecionar os k maiores valores singulares
k = 10  # número de valores singulares que queremos manter
U_k = U[:, :k]
S_k = np.diag(S[:k])
VT_k = VT[:k, :]

# Reconstruir a imagem aproximada
image_approx = np.dot(U_k, np.dot(S_k, VT_k))

# Visualizar a imagem original e a imagem comprimida
plt.subplot(1, 2, 1)
plt.title('Imagem Original')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Imagem Comprimida')
plt.imshow(image_approx, cmap='gray')

plt.show()
```
